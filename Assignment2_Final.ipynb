As shown above, if the file cannot be found by the interpreter or if it cannot be read as a text file, an exception will be raised. Modify function fileLength() so that a friendly message is printed instead:<br>
fileLength('midterm.py')<br>
358<br>

```python
import os;

def fileLength(filename):  
    exists = False
    if filename in os.listdir():
        exists = True

    if exists:
        with open(filename, 'r') as infile:
            content = infile.read()
            return "File {} length = {}".format(filename, len(content))
    else:
        return "File {} not found.".format(filename) # Added statement for Invalid Files


print(fileLength('idterm.py'))       
print(fileLength('midterm.py'))
```

b. supports method jump that takes number values dx and dy as input and moves the kangaroo by dx units along the x-axis and by dy units along the y-axis,<br>
c. overloads the _str_ operator so it behaves as shown below.

```python
# Write a class named Marsupial that can be used as shown below:
class Marsupial:
    def __init__(self):
        self.pouch = []  # Initialize an empty pouch list

    def put_in_pouch(self, item):
        self.pouch.append(item)  # Add the given item to the pouch list

    def pouch_contents(self):
        return self.pouch  # Return the contents of the pouch

# class named Kangaroo as a subclass of Marsupial that inherits all the attributes of Marsupial
# a extends the Marsupial __init__ constructor to take, as input, the coordinates x and y of the Kangaroo object,
class Kangaroo(Marsupial):
    def __init__(self, x=0, y=0):
        super().__init__()  
        self.position = (x, y)  

# b supports method jump that takes number values dx and dy as input and moves the kangaroo by dx units along the x-axis and by dy units along the yaxis, and
    def jump(self, dx, dy):
        x, y = self.position
        self.position = (x + dx, y + dy)  

#  c. overloads the __str__ operator
    def __str__(self):
        x, y = self.position
        return f"I am a Kangaroo located at coordinates ({x}, {y})"  

    

m = Marsupial()
m.put_in_pouch('doll')
m.put_in_pouch('firetruck')
m.put_in_pouch('kitten')
print(m.pouch_contents())            # The result is a list containing the items "doll," "firetruck," and "kitten."

k = Kangaroo(0, 0)
print(k)                             # The value of the variable "k" is printed, and it displays the following message: "I am a Kangaroo located at coordinates (0, 0)."
k.put_in_pouch('doll')
k.put_in_pouch('firetruck')
k.put_in_pouch('kitten')
print(k.pouch_contents())            # The result is a list that includes the elements "doll," "firetruck," and "kitten."
k.jump(1, 0)
k.jump(1, 0)
k.jump(1, 0)
print(k)                             # The message displayed is "I am a Kangaroo positioned at the coordinates (3, 0)."
```

### Question 4
Write function collatz() that takes a positive integer x as input and prints the Collatz sequence starting at x. A Collatz sequence is obtained by repeatedly applying this rule to the previous number x in the sequence:<br>
x = {<br>
ùë•/2 ùëñùëì ùë• ùëñùë† ùëíùë£ùëíùëõ<br>
3ùë• + 1 ùëñùëì ùë• ùëñùë† ùëúùëëùëë<br>
Your function should stop when the sequence gets to number 1. Your implementation must be recursive, without any loops.

```python
def collatz(x):
    sequence = [x]  
    
    while x != 1:  
        if x % 2 == 0:  
            x = x // 2  
        else:  
            x = 3 * x + 1  
        sequence.append(x)  
    
    print(*sequence, sep='\n') # Print the Collatz sequence, separating elements by a new line

collatz(1)
collatz(10)  
```

### Question 5
Write a recursive method binary() that takes a non-negative integer n and prints the binary representation of integer n.<br>

binary(9)<br>
1001<br>

```python
def binary(n):
    if n == 0:  
        print(0, end='')  
        return  

    bits = []  
    while n > 0:  
        bits.insert(0, n % 2)  
        n //= 2  

    print(*bits, sep='', end='')   # Print the list of bits without separators, ending with an empty string

binary(9)
```

### Question 6
Implement a class named HeadingParser that can be used to parse an HTML document, and retrieve and print all the headings in the document. You should implement your class as a subclass of HTMLParser, defined in Standard Library module html.parser. When fed a string containing HTML code, your class should print the headings, one per line and in the order in which they appear in the
document. Each heading should be indented as follows: an h1 heading should have indentation 0, and h2 heading should have indentation 1, etc. Test your implementation using w3c.html.

```python
from html.parser import HTMLParser  
import re  

class HeadingParser(HTMLParser):
    def __init__(self):
        super().__init__()    # Call the superclass (HTMLParser) constructor
        self.indentation = 0  # Initialize the indentation level for headings
        self.headings = []  
        self.in_body = False  # Flag to track if currently parsing within the <body> tag

    def handle_starttag(self, tag, attrs):
        if tag.lower().startswith('h'):                      
            match = re.match(r'h(\d+)', tag, re.IGNORECASE)  # Use regular expression to match the heading level
            if match:  
                self.indentation = int(match.group(1)) - 1   # Set the indentation level based on the heading level
        elif tag.lower() == 'body':  
            self.in_body = True  
            
    def handle_endtag(self, tag):
        if tag.lower() == 'body':  
            self.in_body = False  

    def handle_data(self, data):
        if self.in_body and self.indentation >= 0:  
            heading = ' ' * self.indentation + data.strip()  # Create the heading with appropriate indentation
            if heading:  
                self.headings.append(heading)  

    def print_headings(self):
        for i, heading in enumerate(self.headings):  
            print(heading, end='')  #
            if i != len(self.headings) - 1:  
                print()  

with open('w3c.html') as infile:  
    content = infile.read()  
    
infile.close()
hp = HeadingParser()      # Create an instance of the HeadingParser class
hp.feed(content)          # Parse the HTML content
hp.print_headings()  
```

### Question 7

Implement recursive function webdir() that takes as input: a URL (as a string) and non-negative integers depth and indent. Your function should visit every web page reachable from the starting URL web page in depth clicks or less, and print each web page's URL. As shown below, indentation, specified by indent, should be used to indicate the depth of a URL.<br>

webdir('http://reed.cs.depaul.edu/lperkovic/csc242/test1.html', 2, 0)<br>
http://reed.cs.depaul.edu/lperkovic/csc242/test1.html<br>
___http://reed.cs.depaul.edu/lperkovic/csc242/test2.html<br>
____http://reed.cs.depaul.edu/lperkovic/csc242/test4.html<br>
___http://reed.cs.depaul.edu/lperkovic/csc242/test3.html<br>
____http://reed.cs.depaul.edu/lperkovic/csc242/test4.html<br>

```python
import requests
import re

def webdir(url, depth, indent):          # user defined method with 3 attributes
    print(' ' * indent + url)

    if depth == 0:
        return

    try:
        reply = requests.get(url)        # To access the URL 
        if reply.status_code == 200:
            information = reply.text     # To get accessed content as text

            web_addresses = extract_web_addresses(information)

            for link in web_addresses:
                webdir(link, depth - 1, indent + 4)
    except requests.exceptions.RequestException as e:
        print('Error accessing URL:', url)
        print('Error message:', str(e))

def extract_web_addresses(information):
    url_pattern = r'http[s]?://(?:[a-zA-Z]|[0-9]|[$-_@.&+]|[!*\\(\\),]|(?:%[0-9a-fA-F][0-9a-fA-F]))+' # Specifying pattern for URL
    web_addresses = re.findall(url_pattern, information)
    return web_addresses

webdir('https://www.hotstar.com/ca', 2, 0)
```

### Question 8

Write SQL queries on the below database table that return:<br>

a) All the temperature data.<br>
b) All the cities, but without repetition.<br>
c) All the records for India.<br>
d) All the Fall records.<br>
e) The city, country, and season for which the average rainfall is between 200 and 400 millimeters.<br>
f) The city and country for which the average Fall temperature is above 20 degrees, in increasing temperature order.<br>
g) The total annual rainfall for Cairo.<br>
h) The total rainfall for¬†each¬†season.<br>

```python
import sqlite3

# Establish a connection to the SQLite database, creating a new database if it does not already exist.
conn = sqlite3.connect('weather2.db')
cursor = conn.cursor()

# Generate the table structure.
cursor.execute('''Create Table weather (
                    City TEXT,
                    Country TEXT,
                    Season TEXT,
                    Temperature REAL,
                    Rainfall REAL
                )''')

# Add the data into the table.
data = [('Mumbai', 'India', 'Winter', 24.8, 5.9),
        ('Mumbai', 'India', 'Spring', 28.4, 16.2),
        ('Mumbai', 'India', 'Summer', 27.9, 1549.4),
        ('Mumbai', 'India', 'Fall', 27.6, 346.0),
        ('London', 'United Kingdom', 'Winter', 4.2, 207.7),
        ('London', 'United Kingdom', 'Spring', 8.3, 169.6),
        ('London', 'United Kingdom', 'Summer', 15.7, 157.0),
        ('London', 'United Kingdom', 'Fall', 10.4, 218.5),
        ('Cairo', 'Egypt', 'Winter', 13.6, 16.5),
        ('Cairo', 'Egypt', 'Spring', 20.7, 6.5),
        ('Cairo', 'Egypt', 'Summer', 27.7, 0.1),
        ('Cairo', 'Egypt', 'Fall', 22.2, 4.5)]

cursor.executemany('INSERT INTO weather VALUES (?, ?, ?, ?, ?)', data)

# Finalize the changes and terminate the connection.
conn.commit()
conn.close()
```

```python
# a) All the temperature data.
cursor.execute("SELECT Temperature FROM weather")
temperature_data = cursor.fetchall()
print("All the temperature data:")
for entry in temperature_data:
    print(entry[0])
```

```python
#b) All the cities, but without repetition.
cursor.execute("SELECT DISTINCT City FROM weather")
cities = cursor.fetchall()
print("All the cities without repetition:")
for entry in cities:
    print(entry[0])
```

```python
#c) All the records for India.
cursor.execute("SELECT * FROM weather WHERE Country = 'India'")
records = cursor.fetchall()
print("All the records for India:")
for entry in records:
    print(entry)
```

```python
#d) All the Fall records.
cursor.execute("SELECT * FROM weather WHERE Season = 'Fall'")
fall_records = cursor.fetchall()
print("All the Fall records:")
for entry in fall_records:
    print(entry)
```

```python
#e) The city, country, and season for which the average rainfall is between 200 and 400 millimeters.
cursor.execute("SELECT City, Country, Season FROM weather WHERE Rainfall BETWEEN 200 AND 400")
average_rainfall = cursor.fetchall()
print("City, Country, and Season with average rainfall between 200 and 400 millimeters:")
for entry in average_rainfall:
    print(entry)
```

```python
#f) The city and country for which the average Fall temperature is above 20 degrees, in increasing temperature order.
cursor.execute("SELECT City, Country FROM weather WHERE Season = 'Fall' AND Temperature > 20 ORDER BY Temperature ASC")
average_fall_temperature = cursor.fetchall()
print("City and Country with average Fall temperature above 20 degrees (in increasing temperature order):")
for entry in average_fall_temperature:
    print(entry)
```

```python
#g) The total annual rainfall for Cairo.
cursor.execute("SELECT SUM(Rainfall) FROM weather WHERE City = 'Cairo'")
total_rainfall = cursor.fetchone()[0]
print("The Total annual rainfall for Cairo:", total_rainfall)
```

```python
#h) The total rainfall for each season.
cursor.execute("SELECT Season, SUM(Rainfall) FROM weather GROUP BY Season")
seasonal_rainfall = cursor.fetchall()
print("The total rainfall for each season:")
for entry in seasonal_rainfall:
    print(entry)
```
